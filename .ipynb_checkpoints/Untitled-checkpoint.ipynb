```python
from sympy import *
init_printing()
import numpy as np
from scipy import interpolate

x, y = symbols('x y')
f = sin(x)*sin(y)/(x*y)
f = lambdify((x, y), f)

xs, ys = np.mgrid[-np.pi:np.pi:31j, -np.pi:np.pi:31j]
zs = f(xs, ys)

# observe the nan values, due to division by zero
# these represent places where the given function is undefined
# we'll interpolate these then color them uniquely to indicate
# that these areas of the plot are undefined
zs
```

```python
xx, yy = np.meshgrid(np.arange(0, zs.shape[1]),
                     np.arange(0, zs.shape[0]))

#mask invalid values
zs_masked = np.ma.masked_invalid(zs)

zs_masked
```

```python
#get only the valid values
x1 = xx[~zs_masked.mask]
y1 = yy[~zs_masked.mask]
```

```python
zs_interp = zs_masked[~zs_masked.mask]
zs_interp
```

```python
GD1 = interpolate.griddata((x1, y1), zs_interp.ravel(),
                          (xx, yy),
                             method='cubic')
```
